```python
delay()

quarters = pd.DataFrame()

#Locate table element containing League Player Stats
table0 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="boxscore"]/table')))

# Get the table HTML content
table_html0 = table0.get_attribute('outerHTML')

# Use pandas to parse the HTML table
dfs0 = pd.read_html(table_html0)

# Select the table of interest (you may need to adjust the index)
table_data0 = dfs0[0]

# Append the table data to the main dataframe
game_data = pd.concat([quarters, table_data0], ignore_index=True)
columns_to_exclude = ['Unnamed: 0', 'Unnamed: 7']
game_data = game_data.drop(columns=columns_to_exclude)
game_data = game_data.rename(columns={'Unnamed: 1': 'Team'})

game_data
```

Play-by-play

```python
delay()
delay()
delay()

# Create empty lists to store the scraped data
descriptions = []
teams = []
times = []

# Locate the button element
button1 = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="tab-playbyplay"]'))).click()

delay()
# Wait for the element to be visible and retrieve its text
# Wait for the elements to be present and retrieve them
action = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'pbp__timeline__description'))
)

for text_element in action:
    descriptions.append(text_element.text)
    
```

```python
delay()

team_action = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'pbp__timeline__team'))
)

for text_element in action:
    try:
        descriptions.append(text_element.text)
    except StaleElementReferenceException:
        # Refresh the elements
        action = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, 'pbp__timeline__description'))
        )
        # Retry accessing the element
        descriptions.append(action[text_element].text)
```

```python
delay()

time_in_game = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'pbp__timeline__clock'))
)

for text_element in time_in_game:
    try:
        times.append(text_element.text)
    except StaleElementReferenceException:
        continue


# Close the WebDriver
driver.quit()
```

```python
# Create a dataframe from the scraped data
data = {
    'Description': descriptions,
    'Time': times,
    'Team': teams
}
df = pd.DataFrame(data)

```
