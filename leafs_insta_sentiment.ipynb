```python
#Load packages into python
import pandas as pd
from textblob import TextBlob
import numpy as np
import re
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

#Load file of @mapleleafs instagram comments 
df = pd.read_csv(r'/Users/tmanett/Downloads/Leafs_Comments.csv')
print(df)

```

```python
#filter dataframe for columns needed
comments_df = pd.DataFrame(df, columns = ['timestamp', 'username', 'profileUrl', 'comment', 'likeCount', 'replyCount'])

#function for cleaning comments
def cleanTwt(twt):
    twt = re.sub('#bitcoin', 'bitcoin', str(twt))
    twt = re.sub('#Bitcoin', 'Bitcoin', str(twt))
    twt = re.sub('#[A-za-z0-9]+', '',str(twt))
    twt = re.sub('\\n', '',str(twt))
    twt = re.sub('https?:\/\/S+', '', str(twt))
    return twt

#apply fucntion to 'comment' columns, and name it 'clean_text'
comments_df['clean_text'] = comments_df['comment'].apply(cleanTwt)
comments_df
```

```python
#Apply subjectivity and polarity functions

### Subjectivity refers to the degree to which a statement expresses a personal opinion or feeling, rather than a factual or objective statement. 
###A subjective statement is typically influenced by the author's emotions, biases, and beliefs. In sentiment analysis, subjectivity is often used to distinguish between statements that express a 
###sentiment and those that do not.

###Polarity, on the other hand, refers to the emotional tone of a statement, whether it is positive, negative, or neutral. 
###A statement with positive polarity expresses a favorable sentiment, while a statement with negative polarity expresses an unfavorable sentiment. 
###A neutral statement, by contrast, expresses neither a positive nor a negative sentiment.


def getSubjectivity(twt):
    return TextBlob(twt).sentiment.subjectivity

def getPolarity(twt):
    return TextBlob(twt).sentiment.polarity

    

comments_df['Subjectivity'] = comments_df['clean_text'].apply(getSubjectivity)
comments_df['Polarity'] = comments_df['clean_text'].apply(getPolarity)

comments_df = comments_df.dropna()
comments_df
```

```python
#Get sentiment text
def getSentiment(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'


#Create new column to store text sentiment
comments_df['Sentiment'] = comments_df['Polarity'].apply(getSentiment)
comments_df
```

```python
comments_df.to_csv('Leafs_Comments.csv')
```
