```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv("C:/Users/tmanett/Desktop/Uprising/2k_data.csv")
df
```

```python
weights = {
    'REB/G': 0.05,
    'STL/G': 0.35,
    'DREB': 0.05,
    'DREB%': 0.05,
    'BLK/G': 0.1,
    'STL%': 0.2,
    'BLK%': 0.2
}

# Calculate the league averages for each column
numeric_columns = df.columns.difference(['Player', 'TeamName', 'GPFilter comparator>=Select GP0510152025', 'WinsFilter comparator>=Select Wins0510152025'])  # Exclude non-numeric columns

# Normalize data
normalized_data = (df[numeric_columns] - df[numeric_columns].mean()) / df[numeric_columns].std()
```

```python
# Calculate player scores using weighted sum
player_scores = (normalized_data * pd.Series(weights)).sum(axis=1)

# Add player scores to the DataFrame
df['PlayerScore'] = player_scores

# Sort players by their scores
df = df.sort_values(by='PlayerScore', ascending=False)

df
```

```python
# Filter data for 'Raptors Uprising' team
raptors_df = df[df['TeamName'] == 'Raptors Uprising GC']
```

```python
from itertools import combinations
# Get all possible player combinations for retaining 3 and 4 players
player_combinations_3 = list(combinations(raptors_df['Player'], 3))
player_combinations_4 = list(combinations(raptors_df['Player'], 4))
```

```python
# Initialize dictionaries to store results
results_3_players = {}
results_4_players = {}

# Calculate cumulative scores for each combination of 3 players
for combo in player_combinations_3:
    selected_players = raptors_df[raptors_df['Player'].isin(combo)]
    cumulative_score = (selected_players['PlayerScore']).sum()
    results_3_players[combo] = cumulative_score

# Calculate cumulative scores for each combination of 4 players
for combo in player_combinations_4:
    selected_players = raptors_df[raptors_df['Player'].isin(combo)]
    cumulative_score = (selected_players['PlayerScore']).sum()
    results_4_players[combo] = cumulative_score

# Print results for retaining 3 players
print("Results for retaining 3 players:")
for combo, score in results_3_players.items():
    print(f"{combo}: {score}")

# Print results for retaining 4 players
print("\nResults for retaining 4 players:")
for combo, score in results_4_players.items():
    print(f"{combo}: {score}")
```

```python
# Rank combinations based on cumulative scores
ranked_3_players = sorted(results_3_players.items(), key=lambda x: x[1], reverse=True)
ranked_4_players = sorted(results_4_players.items(), key=lambda x: x[1], reverse=True)

# Print ranked results for retaining 3 players
print("Ranked results for retaining 3 players:")
for rank, (combo, score) in enumerate(ranked_3_players, start=1):
    print(f"{rank}. {combo}: {score}")

# Print ranked results for retaining 4 players
print("\nRanked results for retaining 4 players:")
for rank, (combo, score) in enumerate(ranked_4_players, start=1):
    print(f"{rank}. {combo}: {score}")
```
